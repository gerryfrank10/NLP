# Working With Pipeline



```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!",
    ]
)
```

# Zero-shot classification

```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification')
classifier('his course is transformer course', candidate_labels=['education', 'politics', 'business'])
```

# Text Generation

```python
from transformers import pipeline
generator = pipeline('text-generation')
generator('Hello I have been studying this course for over ')
```

## Using any model in the hub

```python
from transformers import pipeline
generator = pipeline('text-generation', model='distilgpt2')
generator('Our Father who are in heaven', max_length=50, num_return_sequences=3)
```

# Mask Filling

```python
from transformers import pipeline
unmasker = pipeline('fill-mask')
unmasker('And God said to <mask> take of your shoes for the ground is Holy')
```

# Named Entity Recognition

Named Entity is where model has to find what the input text corresponds to entities

```python
from transformers import pipeline
ner = pipeline('ner', grouped_entities=True)
ner('My Name is Gerald and I am student at Bournemouth University')
```

 The model perfectly identifies `Gerald` as the person and `Bournemouth University` as the Organization

# Question-Answering

```python
from transformers import pipeline

qna = pipeline('question-answering')
qna(question='Where do I study', context='My name is Gerald and I am student at Bournemouth University')
```
